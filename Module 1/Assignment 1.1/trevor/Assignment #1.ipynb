A. A Simple Toy Model

Solving the Lane-Emden eqaution for a white dwarf star

```python
from numpy import *
from matplotlib.pyplot import *
import phy4910 as phy


def f(x,y,z):                                 #Function for first derivative
    return z

def g(x,y,z):                                 #Function for second derivative
    return -y**n - 2*z/x

def dimMass(x,y,z):
    return (y**n) * (x**2)

k_nr = 3.166 * 10**12
n = 1.5                                       #for a white dwarf
rho_c = 4.046*10**6                           #g/cm^3


G = 6.674 * 10**-8                            #cm^3 * g^−1 * s^−2
M0 = 1.989 * 10**33                           #g   1 Solar mass
R0= 6.955 * 10**10                            #cm   1 Solar radius



dx = 0.01                                     #integration step size
y0 = 1                                        
x0 = 10**-6
xf = 3.66                                     #The surface is at eta = 3.66, found through trial and error
z0 = 0
X,Y,Z = phy.ode_rk4(x0,xf,dx,y0,z0,f,g)  


phy.plot(X,Y, xlabel ='eta', ylabel = 'varrho' )


print('Surface at eta_s =',xf)



lambda_n = (((n+1)*k_nr*rho_c**((1-n)/n))/(4*pi*G))**(1/2)

m = phy.trapInt(dimMass(X,Y,Z),dx)             #integrate the dimensionless mass
M = 4*pi* lambda_n**3  *rho_c*m[-1]            #calculate the total mass

r = multiply(X,lambda_n)
rho = multiply(Y,rho_c)

#M2 = - 4 * pi * lambda_n**3 * rho_c * xf**2 *Z[-1]   #Calculates total mass with alternate method

phy.plot(X,m ,xlabel = 'eta', ylabel = 'm')

print('Dimensionless Mass, m =',m[-1])
print()
print()
print()
print('Central Density =',format(rho_c,'e'),'g/cm^3')
print('Total mass =', format(M,'e'), 'g')
print('Surface Radius =',format(r[-1]/R0,'e'),'R0')

phy.plot(r,rho,xlabel='Radius (cm)',ylabel='Density(g/cm^3)')
```

The central density that will give a star with the same mass as 1 solar unit is 4.046*10^6 g/cm^3. This produces a star with a radius of 1.277*10^-2 solar radii. 
Changing the central density changes the total mass of the star. The total mass of the star will increase when the central density increases.
The radius of the star also changes when the central density changes. The radius is inversely proportional to the central density, decreasing if central density increases. 
The radius is also inversely proportional to the mass, increasing when mass decreases.

```python
from numpy import *
from matplotlib.pyplot import *
import phy4910 as phy


def f(x,y,z):                                 #Function for first derivative
    return z

def A(x):
    return -(5/9) * x**(-4/3) * (1+x**(2/3))**(-1/2) - (2/3)* x**(-2/3) * (1+x**(2/3))**(-3/2) + (1/3) * (1+x**(2/3))**(-5/2)

def B(x):
    return (5/3) * x**(-1/3) * (1+x**(2/3))**(-1/2) - (1/3)* x**(1/3) * (1+x**(2/3))**(-3/2)

def g(x,y,z):                                 #Function for second derivative
    return -2/x * z - A(y)/B(y)*z**2 - (1/B(y))*y

def dM(x,y):
    return y * x**2

def stop(xi,yi,zi):
    if y0 * 0.001 > yi:
        return True


k_r = 4.936 * 10**14
k_nr = 3.166 * 10**12

rho_0 = 3.789 *10**6                            #g/cm^3

G = 6.674 * 10**-8                              #cm^3 * g^−1 * s^−2
M0 = 1.989 * 10**33                             #g   1 Solar mass
R0= 6.955 * 10**10                              #cm   1 Solar radius


min_rho_c = 4                                   # 10**4 rho_0
max_rho_c = 12                                  # 10**12 rho_0
numModels = 25



rho_cList = logspace(4,12,numModels)


dx = 0.00001
x0 = 10**-6
xf = 20
z0 = 0


Lambda = (k_nr/(4*pi*G*rho_0**(1/3)))**(1/2)

print ('Lambda =' , format(Lambda,'e'))

dr = Lambda * dx

XList =[]
YList =[]
ZList =[]
rList =[]
rSList =[]
rhoList =[]
MList =[]
M_totList =[]

N=len(rho_cList)

for i in range(N):
    
    print('Working...',str(i+1)+'/'+str(N))
    y0=rho_cList[i]/rho_0    
    
    X,Y,Z,Xf = phy.ode_rk4(x0,xf,dx,y0,z0,f,g,stop,return_xf = True)
    XList.append(X)
    YList.append(Y)
    ZList.append(Z)
    
    r = multiply(X,Lambda)
    rho = multiply(Y,rho_0)
    rS = Xf * Lambda

    
    rList.append(r)
    rSList.append(rS)
    
    rhoList.append(rho)
    
    #M2 = multiply(4*pi*(Lambda**3)*rho_0,phy.trapInt(dM(X,Y),dx))     #alternate method to calculate mass
    
    M = multiply(4*pi,phy.trapInt(dM(r,rho),dr))
    
    M_tot = M[-1]
    MList.append(M)
    M_totList.append(M_tot)
    
print('Central Densities')
for i in range(len(rho_cList)):
    print ('Model # '+str(i+1),format(rho_cList[i],'e'),'rho0',format(rho_cList[i]* rho_0,'e'),'g/cm^3' )

print('Total Masses')
for i in range(len(M_totList)):
    print('Model # '+str(i+1),format(M_totList[i],'e'),'g',format(M_totList[i]/M0,'e'),'M0    (Solar Radius)')

print('Surface Radii')
for i in range(len(rSList)):
    print('Model # '+str(i+1),format(rSList[i],'e'),'cm',format(rSList[i]/(R0/100),'e'),'R0/100   (1/100 Solar Radius)')
    
phy.plot(rho_cList,divide(M_totList,M0), xlabel ='Central Density g/cm^3',ylabel='Total Mass M0 (Solar Radius)', logx=True)

phy.plot(rho_cList,divide(rSList,R0/100), xlabel ='Central Density g/cm^3',ylabel='Stellar Radius R0/100 (1/100 Solar Radius)', logx=True)

phy.plot(divide(M_totList,M0), divide(rSList,R0/100), xlabel='Total Mass M0 (Solar Radius)',ylabel='Stellar Radius R0/100 (1/100 Solar Radius)')
```

The first plot shows that as the central density of the star increases the mass also increases, but as it increases the slope flattens out and the masses approach the mass of about 1.4 solar masses. As you might expect this shows the combined effects of the nonrelativistic gas where mass increased with density and the relativistic gas where mass was always 1.4 solar masses. The maximum stable mass is 1.4 solar masses.

The second plot shows that as the central density of the star increases the radius decreases. This is the same relationship we found for both the nonrelativistic and relativistic cases.


The final plot shows that as the mass of the star increases the radius of the star decreases, as it did in the nonrelativstic case. The famous result from this plot is the chandrasekhar limit of white dwarf stars. This is the largest possible mass a white dwarf star can reach before the core collapses under the pressure. The plot shows the limit to be about 1.4 solar masses which is in agreement with the accepted value.
